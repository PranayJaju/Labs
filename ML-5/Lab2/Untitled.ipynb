```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

```python
x = np.arange(0,101)

fig = plt.figure()
ax = fig.add_axes([0,0,1,1])
ax.plot(x,x*2)
ax.set_xlabel('x')
ax.set_ylabel('y')
ax.set_title('title')
plt.locator_params(axis='y', nbins=5)
plt.xlim((0,max(x)))
plt.ylim((0,max(x*2)))
```

```python
x = np.random.randint(0,100,10)
x = np.sort(x)

fig = plt.figure()

ax1 = fig.add_axes([0,0,1,1])
ax1.plot(x, np.exp(x))

ax2 = fig.add_axes([0.2,0.5,.2,.2])
ax2.plot(np.sin(x),x)

fig.show()
```

```python
x = company['month_number']
y = company['total_profit']

fig = plt.figure()

ax = fig.add_axes([0,0,1,1])
ax.set_title('Monthly Profits')
ax.set_xlabel('Month Number')
ax.set_ylabel('Sold Units Number')
ax.plot(x, y, color='red', linestyle='dotted', marker='o', linewidth='3', label='total_profits')
ax.legend(loc='lower right')

fig.show()
```
