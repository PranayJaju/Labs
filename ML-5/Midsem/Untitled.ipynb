```python
import numpy as np
import pandas as pd
```

```python
data = pd.read_csv('data.csv')
```

```python
data
```

```python
def MNB(ho, st, inc):
    
    def_no = len(data[data['Defaulted'] == 'No'])
    def_yes = len(data[data['Defaulted'] == 'Yes'])
    
    p_ho_Dy = len(data[data['HomeOwner'] == ho][data[data['HomeOwner'] == ho]['Defaulted'] == 'Yes'])/def_yes
    p_ho_Dn = len(data[data['HomeOwner'] == ho][data[data['HomeOwner'] == ho]['Defaulted'] == 'No'])/def_no
    
    p_st_Dy = len(data[data['Status'] == st][data[data['Status'] == st]['Defaulted'] == 'Yes'])/def_yes
    p_st_Dn = len(data[data['Status'] == st][data[data['Status'] == st]['Defaulted'] == 'No'])/def_no
    
    p_inc_Dy = len(data[data['Income'] == inc][data[data['Income'] == inc]['Defaulted'] == 'Yes'])/def_yes
    p_inc_Dn = len(data[data['Income'] == inc][data[data['Income'] == inc]['Defaulted'] == 'No'])/def_no
    
    
    print(f'Intermediate Probabilites:')
    print(f'Prob of HomeOwner = {ho} given Def = No: {p_ho_Dn: .3}')
    print(f'Prob of HomeOwner = {ho} given Def = Yes: {p_ho_Dn: .3}')
    print(f'Prob of Status = {st} given Def = No: {p_st_Dn: .3}')
    print(f'Prob of Status = {st} given Def = Yes: {p_st_Dy: .3}')
    print(f'Prob of Income = {inc} given Def = No: {p_inc_Dn: .3}')
    print(f'Prob of Income = {inc} given Def = Yes: {p_inc_Dy: .3}')
    
    prob_def_no = p_ho_Dn*p_st_Dn*p_inc_Dn
    prob_def_yes = p_ho_Dy*p_st_Dy*p_inc_Dy
    
    print(f'\nFinal Probabilites:')
    print(f'Prob of Defaulted = No: {prob_def_no: .3}')
    print(f'Prob of Defaulted = Yes: {prob_def_yes: .3}')
    
    if prob_def_yes > prob_def_no:
        print(f'\nDefaulted: Yes')
    else:
        print(f'\nDefaulted: No')
```

```python
MNB('No', 'Business', 'High')
```
